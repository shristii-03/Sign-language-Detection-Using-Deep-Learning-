```python
# ======= Corrected Colab cell (copy-paste the whole block) =======
# Upload kaggle.json when prompted
from google.colab import files
print("Please upload kaggle.json when prompted.")
files.upload()

# place kaggle.json and set permissions
!mkdir -p ~/.kaggle
!mv -f kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json

# download dataset non-interactively (use --force to avoid overwrite prompt)
print("Downloading dataset (non-interactive)...")
!kaggle datasets download -d grassknoted/asl-alphabet --force -q
print("Unzipping dataset...")
!unzip -qo asl-alphabet.zip -d asl_alphabet_dataset
print("Unzip complete.")

# show GPU to confirm environment
print("GPU info (nvidia-smi):")
!nvidia-smi

# =================== Python imports & sanity ====================
import importlib, sys
# ensure we are using stdlib os (avoid conflicts if there's a user file named os.py)
os = importlib.import_module('os')
import random, json, datetime
import numpy as np
import matplotlib.pyplot as plt
import shutil
%matplotlib inline

# =================== Path detection ====================
base_dir = "/content/asl_alphabet_dataset"
print("Top-level /content (first 120 entries):", os.listdir("/content")[:120])

if not os.path.exists(base_dir):
    raise FileNotFoundError(f"Expected dataset dir at {base_dir} not found. Check unzip step.")

print("Contents of", base_dir, "->", os.listdir(base_dir))

# Candidate paths to search for A/B/C... subfolders
possible_train_paths = [
    os.path.join(base_dir, "asl_alphabet_train"),
    os.path.join(base_dir, "asl_alphabet_train", "asl_alphabet_train"),
    os.path.join(base_dir, "asl_alphabet"),
    base_dir
]

source_root = None
for path in possible_train_paths:
    try:
        if os.path.exists(path) and os.path.isdir(path):
            # find subfolders
            subfolders = [d for d in os.listdir(path) if os.path.isdir(os.path.join(path, d))]
            if len(subfolders) > 0 and any(
                any(fname.lower().endswith(('.jpg', '.jpeg', '.png')) for fname in os.listdir(os.path.join(path, d)))
                for d in subfolders
            ):
                source_root = path
                break
    except Exception as e:
        print(f"Warning inspecting {path}: {e}")

if source_root is None:
    print("Could not auto-detect class subfolders. Directory listing of base_dir:")
    print(os.listdir(base_dir))
    raise FileNotFoundError("Please inspect the unzip layout. A correct layout contains folders A, B, C, ... inside a train folder.")

print("Using source_root =", source_root)

# =================== Hyperparameters (fast test) ====================
IMG_SIZE = (128, 128)
BATCH_SIZE = 32
EPOCHS = 3          # quick run
TRAIN_RATIO = 0.8   # 80% train, 20% validation

# =================== Data generators (use validation_split, no copying) ====================
from tensorflow.keras.preprocessing.image import ImageDataGenerator

val_split = 1.0 - TRAIN_RATIO
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=val_split,
    rotation_range=20,
    width_shift_range=0.12,
    height_shift_range=0.12,
    shear_range=0.12,
    zoom_range=0.12,
    fill_mode='nearest'
)
val_datagen = ImageDataGenerator(rescale=1./255, validation_split=val_split)

seed = 123
train_gen = train_datagen.flow_from_directory(
    source_root,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True,
    seed=seed
)
val_gen = val_datagen.flow_from_directory(
    source_root,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False,
    seed=seed
)

NUM_CLASSES = len(train_gen.class_indices)
print("Classes found:", NUM_CLASSES)
print(train_gen.class_indices)

# =================== Mixed precision (optional speedup on T4) ====================
try:
    from tensorflow.keras import mixed_precision
    mixed_precision.set_global_policy('mixed_float16')
    print("Mixed precision policy:", mixed_precision.global_policy())
except Exception as e:
    print("Mixed precision not enabled (continuing):", e)

# =================== Model (MobileNetV2 backbone) ====================
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models, optimizers

base = MobileNetV2(weights='imagenet', include_top=False, input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
base.trainable = False

x = base.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(256, activation='relu')(x)
x = layers.Dropout(0.4)(x)
# ensure final activation uses float32 to avoid mixed-precision metric issues
outputs = layers.Dense(NUM_CLASSES, activation='softmax', dtype='float32')(x)

model = models.Model(inputs=base.input, outputs=outputs)
opt = optimizers.Adam(1e-3)
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# =================== Callbacks & Train (quick) ====================
from tensorflow.keras import callbacks
logdir = "/content/logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
cbks = [
    callbacks.ModelCheckpoint("/content/sign_model.h5", save_best_only=True, monitor='val_loss'),
    callbacks.EarlyStopping(patience=4, restore_best_weights=True, monitor='val_loss'),
    callbacks.TensorBoard(log_dir=logdir)
]

print(f"Starting training for {EPOCHS} epochs (quick test). Expected ~8-12 minutes on T4.")
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS,
    callbacks=cbks,
)

# =================== Fine-tune (SKIPPED by default for speed) ====================
FT_EPOCHS = 0   # set to >0 if you want to run fine-tuning later
history_ft = None
if FT_EPOCHS > 0:
    print("Starting fine-tuning (this will add time)...")
    base.trainable = True
    for layer in base.layers[:-30]:
        layer.trainable = False
    model.compile(optimizer=optimizers.Adam(1e-5), loss='categorical_crossentropy', metrics=['accuracy'])
    history_ft = model.fit(train_gen, validation_data=val_gen, epochs=FT_EPOCHS, callbacks=cbks)

# =================== Plotting helper ====================
def plot_hist(h1, h2=None):
    acc = h1.history.get('accuracy', []) + (h2.history.get('accuracy', []) if h2 else [])
    val_acc = h1.history.get('val_accuracy', []) + (h2.history.get('val_accuracy', []) if h2 else [])
    loss = h1.history.get('loss', []) + (h2.history.get('loss', []) if h2 else [])
    val_loss = h1.history.get('val_loss', []) + (h2.history.get('val_loss', []) if h2 else [])
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1); plt.plot(acc, label='train_acc'); plt.plot(val_acc, label='val_acc'); plt.legend(); plt.title('Accuracy')
    plt.subplot(1,2,2); plt.plot(loss, label='train_loss'); plt.plot(val_loss, label='val_loss'); plt.legend(); plt.title('Loss')
    plt.show()

plot_hist(history, history_ft)

# =================== Evaluation: confusion matrix & report ====================
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

val_gen.reset()
probs = model.predict(val_gen, verbose=1)
y_pred = np.argmax(probs, axis=1)
y_true = val_gen.classes
labels = list(train_gen.class_indices.keys())

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted'); plt.ylabel('True'); plt.title('Confusion Matrix'); plt.show()

print("Classification Report:\n")
print(classification_report(y_true, y_pred, target_names=labels, digits=3))

# =================== Save artifacts & write local demo script ====================
label_map = {v:k for k,v in train_gen.class_indices.items()}
with open('/content/label_map.json','w') as f:
    json.dump(label_map, f)
print("Saved /content/label_map.json")
if os.path.exists('/content/sign_model.h5'):
    print("Saved /content/sign_model.h5")

webcam_script = f"""
# local_webcam_infer.py (run locally)
import cv2, numpy as np, json
from tensorflow.keras.models import load_model

model = load_model('sign_model.h5')
with open('label_map.json','r') as f: label_map = json.load(f)
IMG_SIZE = {IMG_SIZE}

cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print('Could not open webcam'); exit()
while True:
    ret, frame = cap.read()
    if not ret: break
    img = cv2.resize(frame, IMG_SIZE)
    arr = img.astype('float32')/255.0
    pred = model.predict(np.expand_dims(arr,0))[0]
    idx = int(pred.argmax()); label = label_map.get(str(idx), label_map.get(idx,'Unknown'))
    conf = pred[idx]
    cv2.putText(frame, f'{{label}} ({{conf:.2f}})', (10,30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
    cv2.imshow('Webcam - q to quit', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'): break
cap.release(); cv2.destroyAllWindows()
"""
with open('/content/local_webcam_infer.py','w') as f:
    f.write(webcam_script)
print("Wrote /content/local_webcam_infer.py — download and run locally with sign_model.h5 and label_map.json")

print("All done ✅")
```
